```python
titanic_df = pd.read_csv ('titanic.csv')
titanic_df.head()
```

```python
titanic_df.columns
```

```python
size_map = {'male':1, 'female':2}
titanic_df['Sex'] = titanic_df['Sex'].map(size_map)
titanic_df
```

using map we can change sex to either 1 or 0 this makes it easier to read

```python
class_labels = LabelEncoder()
titanic_df['Cabin'] = class_labels.fit_transform(titanic_df['Cabin'].values)
titanic_df
```

This is to create a easier time reading the vlaues for the cabin as now they are just numbered rather than A(x) and most likely won't have a correlation for the survival so the data isn't super important

```python
titanic_dummy = pd.get_dummies(titanic_df[['Embarked']])
titanic_dummy
```

```python
titanic_df = pd.concat([titanic_df, titanic_dummy], axis=1)
titanic_df
```

```python
titanic_df = titanic_df.drop(['Embarked', 'Ticket','Name'], axis = 1)
titanic_df
```

Here we use one hot encoding to see where the embarked and then we can use the drop function to get rid of the regular embarked color and we use concat to combine the two data sets. Also dropped the names and ticket to create the correlation heat map

```python
import time
```

```python
bins = [0,18,35,50,100]
labels = ['child','early adult','adult', 'elder']
pd_cut = time.time()
pd.cut(titanic_df['Age'], bins=bins, labels=labels)
end_cut = time.time()
lambda_time_start = time.time()
titanic_df['age_bin_lambda'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 18 else 'Early Adult' if x <=30 else 'Adult' if x <= 90 else 'Elder')
lambda_time_end = time.time()
print(end_cut-pd_cut)
print(lambda_time_end-lambda_time_start)
```

the lambda function was faster to find the age groups

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 8))
data_child = titanic_df[titanic_df['age_bin_lambda'] == 'Child']
data_young_adult = titanic_df[titanic_df['age_bin_lambda'] == 'Early Adult']
data_adult = titanic_df[titanic_df['age_bin_lambda'] == 'Adult']
data_senior = titanic_df[titanic_df['age_bin_lambda'] == 'Elder']

sns.histplot(data= data_child, x='Fare', ax=axes[0])
sns.histplot(data= data_young_adult, x='Fare', ax=axes[1])
sns.histplot(data= data_adult, x='Fare', ax=axes[2])
sns.histplot(data= data_senior, x='Fare', ax=axes[3])

```

This shows that for childen, there was higher fare costs more frequently and as with young adults, there was the least amount of high costs for a fare. There was a small amount in young adults and a medium sized amount in seniors, but children had the most higher cost fares out of the four.
